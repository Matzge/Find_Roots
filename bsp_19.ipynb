# Vislaulisation of Functions

&rarr;  $p(x) = (x^4 - \frac{x^3}{2} +\frac{x^2}{2} - \frac{x}{2} -\frac{1}{2}) $

&rarr;  $q(x) = sin(\frac{x}{1000})$

&rarr;  $s(x) = \sum_{k=0}^{n} w_k * x^k$



```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import math
roots = []
results = []


x=np.arange(1/1000,1, 0.0000001)
y=np.sin(1/x)


with open ('Nullstellen_QX.txt') as inputfile:
    for row in csv.reader(inputfile):
        roots.append(row[0])
        for i in range(len(roots)):
            roots[i] = float(roots[i])

for i in range(len(roots)):   
    point = np.sin(1/roots[i])
    point = round(point,2)           #I round here because my roots are not exact from my bisect algorithm
    results.append(point)            #The x values only have a precision of EPS = 10^-6.

    
plt.plot(x,y)
plt.axhline(y=0, color='black', linestyle='-')
plt.plot(roots,results, "ro")


plt.xlabel("x")
plt.ylabel("q(x)")
plt.title("Visualisation of roots of q(x)")

print("Roots:",roots)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import math
roots = []
results = []


x=np.arange(-2.5,2, 0.0001)
y = (x**4 - (x**3)/2 +(x**2)/2 -(x/2) -0.5)*np.cos(x)


with open ('Nullstellen_PX.txt') as inputfile:
    for row in csv.reader(inputfile):
        roots.append(row[0])
        for i in range(len(roots)):
            roots[i] = float(roots[i])
            
for i in range(len(roots)):   
    point = (roots[i]**4 - (roots[i]**3)/2 +(roots[i]**2)/2 -(roots[i]/2) -0.5)*np.cos(roots[i])
    point = round(point,2)           #I round here because my roots are not exact from my bisect algorithm
    results.append(point)            #The x values only have a precision of EPS = 10^-6.

    
print("Roots:",roots)

plt.plot(x,y)
plt.axhline(y=0, color='black', linestyle='-')
plt.plot(roots,results, "ro")


plt.xlabel("x")
plt.ylabel("p(x)")
plt.title("Visualisation of roots of p(x)")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import math
consts = []
roots = []
results = []
y_values = []
x_values = []
with open ('input_1.txt') as inputfile:
    for row in csv.reader(inputfile):
        consts.append(row[0])
        for i in range(len(consts)):
            consts[i] = float(consts[i])

def s(x):
    result = 0
    for i in range(len(consts)):
         result+= consts[i] *x**i
    return result


for i in range(-6000,15000):
    y = s(i/1000)
    y_values.append(y)
    x_values.append(i/1000)
    

with open ('Nullstellen_SX.txt') as inputfile:
    for row in csv.reader(inputfile):
        roots.append(row[0])
        for i in range(len(roots)):
            roots[i] = float(roots[i])

for i in range(len(roots)):   
    point = s(roots[i])
    point = round(point,2)           #I round here because my roots are not exact from my bisect algorithm
    results.append(point)            #The x values only have a precision of EPS = 10^-6.
 
    
plt.plot(x_values,y_values)
plt.axhline(y=0, color='black', linestyle='-')
plt.plot(roots,results, "ro")

print("Roots:",roots)

plt.xlabel("x")
plt.ylabel("s(x)")
plt.title("Visualisation of roots of s(x)")

```
